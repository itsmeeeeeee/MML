```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Mount Google Drive
from google.colab import drive
drive.mount('/content/drive', force_remount=True)
```

```python
file_path_labels = "/content/drive/MyDrive/MultimodalNLP/projekt/data/labels.csv"
labeled_data = pd.read_csv(file_path_labels)
labels = labeled_data['overall_sentiment'].values
```

```python


# Label-Mapping definieren
label_mapping = {
    "very_negative": 0,
    "negative": 0,
    "neutral": 2,
    "positive": 1,
    "very_positive": 1
}

# Textlabels zu numerischen Labels konvertieren
numeric_labels = np.array([label_mapping[label] for label in labels])

unique, counts = np.unique(numeric_labels, return_counts=True)
label_counts = dict(zip(unique, counts))

# Farben definieren
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']

```

```python


# Diagramm 1: Ohne neutrale Klasse
plt.subplot(1, 2, 1)
sizes = [label_counts.get(0, 0), label_counts.get(1, 0)]
labels = ['Negative', 'Positive']
plt.pie(sizes, colors=colors[:2], labels=labels, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Binary Classification (Positive/Negative)')

# Diagramm anzeigen
plt.show()
```

```python
# Diagramm 2 Original data:
plt.subplot(1, 2, 2)
sizes = [label_counts.get(0, 0), label_counts.get(1, 0), label_counts.get(2, 0)]
labels = ['Negative', 'Positive', 'Neutral']
plt.pie(sizes, colors=colors, labels=labels, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Class Distribution Multiclass (Original)')


# Diagramm anzeigen
plt.show()
```
